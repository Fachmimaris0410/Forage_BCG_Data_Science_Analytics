## 5. Verify the Hypotesisof Price Sensitivity

Calculating the change in the proportion of monthly costs

```python
df.loc[df['forecast_cons_12m'] == 0, 'forecast_cons_12m'] = np.nan
df.loc[df['cons_12m'] == 0, 'cons_12m'] = np.nan
```

```python
df['monthly_cost_change_pct']=(df['cons_12m'] - df['forecast_cons_12m']) / df['forecast_cons_12m']*100
```

```python
df['monthly_cost_change_pct']
```

Calculate price sensitivity metrics based on the definitions you have defined. For example, you can calculate the average percentage change in monthly fees for churn and non-churn customers.

```python
import numpy as np

# change value 0 with NaN in column 'forecast_cons_12m'
# df.loc[df['forecast_cons_12m'] == 0, 'forecast_cons_12m'] = np.nan

# Calculates the average percentage change in monthly fees for churn customers
average_cost_change_churn = df[df['churn'] == 1]['monthly_cost_change_pct'].mean()

# Calculates the average percentage change in monthly fees for non-churn customers
average_cost_change_not_churn = df[df['churn'] == 0]['monthly_cost_change_pct'].mean()
```

```python
average_cost_change_churn
```

```python
average_cost_change_not_churn
```

Correlation analysis between price sensitivity metrics and the churn column uses statistical methods such as Pearson correlation or Spearman correlation.

```python
# Calculates the Pearson correlation between the price sensitivity metric and the churn column
correlation = df['monthly_cost_change_pct'].corr(df['churn'])
```

```python
correlation 
```

In this case, the correlation value of -0.007657625851998709 which is very close to 0 indicates that there is no significant linear relationship between the price sensitivity metric and the churn column in the dataset. This means that changes in the percentage of monthly fees (price sensitivity) in general do not significantly affect customer churn decisions.
